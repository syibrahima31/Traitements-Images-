## Définition d’une image 


La représentation d’une image se fait sur un support spatial, par exemple à 2 dimensions (pour une photographie),
ou 3 dimensions (pour une image représentée en volume, sur un ordinateur).
La définition d’une image est le nombre de points décrivant l’image sur son support spatial. Ces points sont appelés
pixels. En général, le nombre de pixels en ligne et en colonne est mentionné : par exemple, une image de 640 par 480
pixels.

```python
camera = data.camera()

zoom = camera[100:200, 200:300]
plt.imshow(zoom, cmap="gray")
print(zoom.shape )
```

```python
 for i in [1,2,4,6,10,20]:
   zoom = camera[::i, ::i]
   plt.imshow(zoom)
   plt.show()
```

## Résolution d’une image

La résolution d’une image ne prend de sens que lors de la représentation de celle-ci sur un support physique, par
exemple un écran d’ordinateur ou un papier photo. Elle se définit par un nombre de points (pixels) par unité de
longueur, par exemple le point par pouce (ppp ou en anglais dpi, dot per inch) : c’est une densité.


Ici la résolution augmente car taille de image devient de plus en plus petit

```python
for s in [1,2,4,5,7,8]:
  plt.figure(figsize=(8/s, 8/s))
  plt.imshow(camera, cmap="gray")
```
