## Importation des packages 

```python
from skimage import data , io 
import matplotlib.pyplot as plt 
```

## Importation d'une image en niveau de gris 

```python
camera = data.camera()

# un seul caneau
camera.shape 

# visualisation de image 
plt.imshow(camera, "gray")
```

## Chargement d'une image couleurs 

```python
chelsea = data.chelsea()

# nombre de canneau : 3 canneaux
print(chelsea.shape)
```

```python
plt.figure(figsize=(7,11))
for i in range(3):
  plt.subplot(3,1,i+1)
  plt.imshow(chelsea[:,:,i], "gray")
```
